# Modelos de Ensamble

###### El principio detras de este modelo es que un grupo debil de aprendices (varias técnicas de aprendizaje automático)  se unan para formar un solo aprendizaje mas robusto con el fin de reducir la varianza, el bias (error) y mejorar las predicciones. Por lo tanto incrementan el accuracy del modelo.

#### Características 

Se dividen en dos grupos: métodos secuenciales(ejemplo: AdaBoost) y métodos paralelos(random forest)

Para el siguiente ejemplo utilizaré Ada Boost

##### Primero debemos definir las librerias a utilizar

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import cross_val_score, train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.3,random_state=13)
```

Crear el AdaBoost classifier object y jugaremos con los valores de learning rate que contribuye a regular los pesos de los weak learners

```python
adac = AdaBoostClassifier(n_estimators=100,
                         learning_rate=2)
```

```python
model = adac.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

Evaluaremos el accuracy de este modelo para nuestro set de datos 

```python
print("Accuracy:",accuracy_score(y_test, y_pred))
```

Como comentarios acerca del modelo, es muy bueno y permite crear un algoritmo mas confiable a mi parecer, porque toma en cuenta todo el aprendizaje de clasificadores para construir uno robusto pero mi conjunto de datos al parecer le hace falta mas limpieza y preprocesamiento por que mi accuracy score es muy bajo.
